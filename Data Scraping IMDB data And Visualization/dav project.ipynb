```python
#importing required Libraries
import pandas as pd   #to create dataframe
import requests       #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML
import numpy as np  # to count the values (in our case)
from time import sleep
from random import randint


# though ihavent used selenium so far, yet i have added the libraries so that if needed, i wont need to import them again and again.
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
```

                PART ONE, SCRAPPING THE IMDB WEBSITE FOR MAKING A DATA SET

```python
#Declaring the headers as we want everything in English 
headers = {"Accept-Language": "en-US,en;q=0.5"}
```

```python
#creating an array of values and passing it in the url for dynamic webpages
pages = np.arange(1,1000,100)
#this way, we will create a list from 1 to 1000 with a step of 100
```

        THESE WILL WORK LIKE OUT ATTRIBUTES (COLUMNS) WHEN WE WILL CONVERT OUR SCRAPPED DATA INTO PD_DATAFRAMES

```python
#creating an empty list, so that we can append the values in them later. There empty lists will also behave like the attributes of the dataframe.
movie_name = []
year = []
time = []
rating = []
metascore = []
votes = []
gross = []
description = []
Director = []
Stars = []
Genre = []
PG_Rating = []


#ORIGINAL LINK TO THE PAGE FROM WHERE I STARTED SCRAPPING FROM SCRATCH.

link = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
```

        HERE WE APPLY 2 FOR LOOPS 
        FIRST LOOP WILL GO THRU EACH AND EVERY WEBPAGE WHERE EACH PAGE CONTAINS 100 MOVIES DATA
        SECOND LOOP WILL GO THRU EACH PAGE AND SCRAPE ALL THE REQUIRED DATA WE WANT FROM THE PAGE

```python
#the whole core of the script
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-advanced'})
    sleep(randint(2,8))
    
#calling one by one using for loop
    for store in movie_data:
        #scrapping the movie data
        name = store.h3.a.text #storing the movie name as it is stored in the h3 tag and a tag
        movie_name.append(name)
        #now every movie name will be appendedd in the list movie_name
    
    
    
        #scrapping the year of release data
        #find function is basically used when you dont know the exact tag name, but you know the attribute name
        year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
        #here i am also replacing the brackets with empty string
        year.append(year_of_release)
        #now every year will be appendedd in the list year
    
    
    
    
        #scrapping the time duration of each movie
        runtime = store.p.find('span', class_ = 'runtime').text.replace(' min', '')
        #here i have replaced the minutes string with an empty string because it is not needed
        time.append(runtime)
        #now every runtime of movie will be stored in the time list
        
        
        #scrapping the parental rating of each movie
        pg_rating = store.p.find('span', class_ = 'certificate').text if store.find('span', class_ = 'certificate') else '---'
        PG_Rating.append(pg_rating)
        
        
    
        #scrapping the genre of each movie
        genres = store.p.find('span', class_ = 'genre').text.replace('\n', '')
        Genre.append(genres)
        
        
    
     #scrapping the rating 
        rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
        #now we have the ratings of the mmovies
        rating.append(rate)
    
    
    
        #finding the metascore
        #right now im not imputing the values of null values in metaphore col, will do it later 
        meta  = store.find('span', class_ = 'metascore').text.replace(' ', '') if store.find('span', class_ = 'metascore') else '-----'
        #there are some movies where metascore is not given, so incase the metascore is not given, i will just fill it with garbage
        metascore.append(meta)
    
    
    
        #since, gross and votes have same attributes,both are in p tag, both have nv,  that's why we had created a common variable and then used indexing
        value = store.find_all('span', attrs = {'name': 'nv'})
    
    
    
        vote = value[0].text
        #now the vote has the first index of the list value which contains both votes and gross
        votes.append(vote)
    
    
    
    
        #where gross is not given for any movie im not imputing it atm. will do it later
        grosses = value[1].text if len(value) >1 else '*****'
        gross.append(grosses)
    
    
    
        # Description of the Movies --  
        describe = store.find_all('p', class_ = 'text-muted')
        description_ = describe[1].text.replace('\n', '') if len(describe) >1 else '*****'
        description.append(description_)
    
    
    
    
        #Cast Details -- Scraping Director name and Stars 
        cast = store.find("p", class_ = '')
        cast = cast.text.replace('\n', '').split('|')
        cast = [x.strip() for x in cast]
        cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
        Director.append(cast[0])
        Stars.append([x.strip() for x in cast[1].split(",")])
        # stars.append([x.strip() for x in cast[1].split(","))
        #

```
